```python
import matplotlib.pyplot as plt
```

```python
# Sample data
genders = ['Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male']
```

```python
# Count the occurrences of each gender
gender_counts = {}
for gender in genders:
    if gender in gender_counts:
        gender_counts[gender] += 1
    else:
        gender_counts[gender] = 1
```

```python
# Specify colors for each gender
colors = ['skyblue', 'pink']
```

```python
# Create bar chart
plt.bar(gender_counts.keys(), gender_counts.values(), color=colors)
# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in Population')
# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate random age data
np.random.seed(0)  # for reproducibility
ages = np.random.randint(18, 90, size=100)  # generate 100 random ages between 18 and 90
```

```python
# Create histogram
plt.hist(ages, bins=10, color='orange', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in Population')

# Show plot
plt.show()
```
